```python
''' 1. Analice los patrones de llegada de los clientes utilizando diferentes distribuciones de probabilidad. '''
"""
Analice los patrones de llegada de los clientes utilizando diferentes distribuciones de probabilidad
1. Simule las llegadas de clientes durante un período de 10 horas mediante una distribución geométrica con
parámetro p=0,2.
2. Grafique los tiempos de llegada de los clientes y calcule el tiempo promedio entre llegadas.
3. ¿Qué modela la distribución geométrica en este escenario?
4. ¿Cómo se compara el tiempo promedio entre llegadas con el valor esperado de la distribución geométrica
"""

import numpy as np
import matplotlib.pyplot as plt

p = 0.2
num_simulations = 1000 

arrival_times = np.random.geometric(p, num_simulations)

# Gráfica de los tiempos de llegada de los clientes
plt.hist(arrival_times, bins=20, density=True, alpha=0.7, color='b')
plt.xlabel('Tiempo entre llegadas')
plt.ylabel('Frecuencia')
plt.title('Distribución de tiempos de llegada de clientes')
plt.grid(True)
plt.show()

# Cálculo del tiempo promedio entre llegadas
mean_arrival_time = np.mean(arrival_times)
print("Tiempo promedio entre llegadas:", mean_arrival_time, "minutos")
```

**¿Qué modela la distribución geométrica en este escenario?**

- Este modela el timpo de llegadas de los clientes en donde la llegada de cada uno es independiente de los demás. El tiempo entre cada llegada se mide en minutos.

**¿Cómo se compara el tiempo promedio entre llegadas con el valor esperado de la distribución geométrica?**

- El valor esperado o media de una distribución geométrica con probabilidad de éxito p está dado por E(X) = 1/p. En este caso, p=0,2, por lo que el valor esperado de la distribución geométrica sería E(X) = 1/0,2 = 5 minutos. El cuál es el mismo tiempo obtenido en el modelo.

```python
''' 2. Modele los tiempos de servicio en un sistema de colas utilizando una distribución binomial negativa '''
r = 5
p = 0.4

print("\n----- Distribucion binomial negativa de una muestra de 500 -----")
Dist_binomial_negativa = np.random.negative_binomial(r, p, size=500)
print(Dist_binomial_negativa)

#Grafico 
print("\n----- Grafico -----")
plt.hist(Dist_binomial_negativa)
plt.title("Histograma de los tiempos de servicio")
plt.xlabel('Tiempos de servicio')
plt.ylabel('Probabilidad')
plt.show()

print("\n----- Varianza -----")
vart = r * (1-p) / (p**2)
var = np.var(Dist_binomial_negativa)

print("Varianza teorica:", vart)
print("Varianza de la muestra:", var)

print("\n----- Media -----")
medt = r * (1-p)/p
med = np.median(Dist_binomial_negativa)

print("Media teorica:", medt)
print("Media de la muestra:", med)
```

**¿Cómo se ajusta la distribución binomial negativa a los datos de tiempo de servicio?**
- Esta distribución se ajusta a los datos de tiempo de servicio ya que deseamos saber el número de ensayos necesarios para que una cierta cantidad de eventos puedan ocurrir y de esta manera se puede preever los tiempos con altas demandas y así poder contar con la capacidad necesaria para atender dicha cantidad de eventos 

**Compare la media y la varianza de los datos con los valores teóricos de la distribución binomial negativa.**
- Como se puede ver los valores son muy simlares y estos solo difieren por decimales lo cual nos puede dar la pauta de que la muestra aleatoria se aproxima a la distribución binomial negativa con los parámetros brindados con aterioridad
