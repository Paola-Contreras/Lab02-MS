1. Suponga que la cantidad promedio de buses que llegan a una parada de bus dada es de 2 cada 30 minutos.
Considere X como la cantidad de buses que llegan a la mencionada parada de bus.

**¿Puede ser este evento modelado por una distribución de Poisson? ¿Por qué?**
- Si, este evento es posible poderlo modelarlo utilizando la distribución de Poisson ya que se desea saber la probabilidad de que llegue un bus a la parada en un intervalo de tiempo especifico. 


```python
'''2. Calcule y grafique la probabilidad para diferentes números de buses, yendo desde 0 hasta 100. ¿Cuál es la 
cantidad de buses más probable? '''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import *

# Cantidad de buses que llegan
lambda_b = 2
x = np.arange(101)
probability = poisson.pmf(x, lambda_b)

print("----- Cantidad de buses mas probable -----\n ", probability)

print("\n----- Grafico -----")
plt.title("Grafico de la cantidad de buses que llegan en 30min")
plt.xlabel("Cantidad de buses")
plt.ylabel("Probabilidad")
plt.hist(x, probability)
plt.tight_layout()

```

2. Asuma que usted trabaja en una industria relacionada con la veterinaria, con lo que sabe que una clínica
determinada el promedio de llegada de pacientes es de 5 por hora

```python
''' 1. Genere una tabla que muestre los tiempos intermedios en horas para los 10 primeros pacientes '''
lambda_p = 5
x = 10

prob = np.random.uniform(size=x)
tiempo_intermedio = -np.log(1 - prob) / lambda_p

print("Paciente\t Tiempo Intermedio")
for i in range(x):
    print(f"{i+1}\t\t {tiempo_intermedio[i]:.6f}")
```

```python
''' 2.Grafique usando el mismo lambda la CDF para la variable exponencial. '''

# Graficar la CDF de la variable exponencial
x = np.linspace(0, 2, 100)
cdf_exponencial = 1 - np.exp(-lambda_p * x)

plt.plot(x, cdf_exponencial)
plt.title("CDF Exponencial")
plt.xlabel("Tiempo Intermedio en horas")
plt.ylabel("Probabilidad")
plt.title("CDF de la Variable Exponencial")
plt.grid(True)
plt.show()
```


3. Considerando que X1, X2, ... Xk son los tiempos intermedios, si definimos T1, T2, ... Tk como las variables que
representarán las llegadas de los pacientes a la clínica veterinaria, vemos que
T1 = X1
T2 = X1 + X2
Tk = X1 + X2 + ... + Xk

**¿Son las variables T1, T2 hasta Tk variables aleatorias? ¿Por qué?**
- Si, estas nuevas variables tambien son aleatorias puesto a que dependen del resultado de la primera variable y por ende se encuentran sujetas a la aleatoridad de esta por lo que los valores de las nuevas variables pueden cambiar a lo largo del evento. 

```python
''' Simule y grafique el proceso de Poisson completo usando el mismo gamma e información recabada del
ejercicio 2. Haga una simulación para 100 pacientes.
'''
# Task 3 pero con 100 
x = 100
uniform_values = np.random.rand(x)
intermediate_times = -np.log(1 - uniform_values) / lambda_p

# Calcular los tiempos de llegada acumulados (proceso de Poisson)
arrival_times = np.cumsum(intermediate_times)

# Graficar el proceso de Poisson
plt.plot(arrival_times, np.arange(1, x + 1))
plt.title("Proceso de Poisson con 100 pacientes")
plt.xlabel('Tiempo en horas')
plt.ylabel('No. pacientes')
plt.show()
```
