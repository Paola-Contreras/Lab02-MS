```python
'''2. Calcule y grafique la probabilidad para diferentes números de buses, yendo desde 0 hasta 100. ¿Cuál es la 
cantidad de buses más probable? '''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import *

# Cantidad de buses que llegan
lambda_b = 2
x = np.arange(101)
probability = poisson.pmf(x, lambda_b)

print("----- Cantidad de buses mas probable -----\n ", probability)

print("\n----- Grafico -----")
plt.title("Grafico de la cantidad de buses que llegan en 30min")
plt.xlabel("Cantidad de buses")
plt.ylabel("Probabilidad")
plt.hist(x, probability)
plt.tight_layout()

```

representarán las llegadas de los pacientes a la clínica veterinaria, vemos que

**¿Son las variables T1, T2 hasta Tk variables aleatorias? ¿Por qué?**
- Si, estas nuevas variables tambien son aleatorias puesto a que dependen del resultado de la primera variable y por ende se encuentran sujetas a la aleatoridad de esta por lo que los valores de las nuevas variables pueden cambiar a lo largo del evento. 

```python
''' Simule y grafique el proceso de Poisson completo usando el mismo gamma e información recabada del
ejercicio 2. Haga una simulación para 100 pacientes.
'''
# Task 3 pero con 100 
x = 100
uniform_values = np.random.rand(x)
intermediate_times = -np.log(1 - uniform_values) / lambda_p

# Calcular los tiempos de llegada acumulados (proceso de Poisson)
arrival_times = np.cumsum(intermediate_times)

# Graficar el proceso de Poisson
plt.plot(arrival_times, np.arange(1, x + 1))
plt.title("Proceso de Poisson con 100 pacientes")
plt.xlabel('Tiempo en horas')
plt.ylabel('No. pacientes')
plt.show()
```

```python
''' 2. Modele los tiempos de servicio en un sistema de colas utilizando una distribución binomial negativa '''
r = 5
p = 0.4

print("\n----- Distribucion binomial negativa de una muestra de 500 -----")
Dist_binomial_negativa = np.random.negative_binomial(r, p, size=500)
print(Dist_binomial_negativa)

#Grafico 
print("\n----- Grafico -----")
plt.hist(Dist_binomial_negativa)
plt.title("Histograma de los tiempos de servicio")
plt.xlabel('Tiempos de servicio')
plt.ylabel('Probabilidad')
plt.show()

print("\n----- Varianza -----")
vart = r * (1-p) / (p**2)
var = np.var(Dist_binomial_negativa)

print("Varianza teorica:", vart)
print("Varianza de la muestra:", var)

print("\n----- Media -----")
medt = r * (1-p)/p
med = np.median(Dist_binomial_negativa)

print("Media teorica:", medt)
print("Media de la muestra:", med)
```

**¿Cómo se ajusta la distribución binomial negativa a los datos de tiempo de servicio?**
- Esta distribución se ajusta a los datos de tiempo de servicio ya que deseamos saber el número de ensayos necesarios para que una cierta cantidad de eventos puedan ocurrir y de esta manera se puede preever los tiempos con altas demandas y así poder contar con la capacidad necesaria para atender dicha cantidad de eventos 

**Compare la media y la varianza de los datos con los valores teóricos de la distribución binomial negativa.**
- Como se puede ver los valores son muy simlares y estos solo difieren por decimales lo cual nos puede dar la pauta de que la muestra aleatoria se aproxima a la distribución binomial negativa con los parámetros brindados con aterioridad

```python
''' 4. En este ejercicio deberá comparar la bondad de ajuste (goodness-of-fit) de diferentes distribuciones de probabilidad
para un conjunto de datos determinado '''

# Muetra de 1000 pts
n = np.random.randn(1000)

#Ajuste por medio de MLE de diferentes distribuciones 
# Dist Normal 
p_normal = stats.norm.fit(n)
dist_normal = stats.norm(*p_normal)

# Dist Binomal
p_exponencial = stats.expon.fit(n)
dist_exponencial = stats.expon(*p_exponencial)

#Graficos
print("\n----- Graficos -----")
plt.title('Ajuste de los datos mediante MLE')
plt.hist(n)
plt.show()

x = np.linspace(np.min(n), np.max(n), 100)
plt.title('Ajuste de Distribuciones mediante MLE')
plt.plot(x, dist_normal.pdf(x), 'r', label='Dist_Normal')
plt.plot(x, dist_exponencial.pdf(x), 'g', label='Dist_Exponencial')
plt.legend()
plt.show()

# Prueba de bondad de ajuste
ks_stat_normal, p_value_normal = kstest(n, 'norm', args=p_normal)
ks_stat_exponencial, p_value_exponencial = kstest(n, 'expon', args=p_exponencial)


print("\n----- Prueba de Bondad Kolmogorov-Smirnov -----")
print("\nKS para la distribución normal:")
print("Estadístico KS:", ks_stat_normal)
print("Valor p:", p_value_normal)

print("\nKS para la distribución exponencial:")
print("Estadístico KS:", ks_stat_exponencial)
print("Valor p:", p_value_exponencial)
```

- Al momento de evaluar los resultados de la prubea de Kolmogorov - Smirnov para dos distribuciones diferentes se puede destacar que en la distribución normal el estadistico no es tan elevado por lo que no hay mucha discrepancia entre los datos empiricos y teoricos, caso contrario para la distrivucion exponencial puesto a que este cuenta con un valor más elevado dandonos a entender que hay una diferencia significativa. Ahora bien al evaluar el valor de p podemos ver que para ambas distribuciones este cuenta con un valor superior a 0.05 lo cual nos indica que no hay suficientes datos para poder ver si la muestra es de una distribucion diferente a la teorica o la muestral
